PREPROCESSING

```python
moviedata= moviedata[['id','title','overview','genres','keywords','cast','director']]
```

```python
moviedata.isnull().sum()
```

```python
moviedata.dropna(inplace=True)
```

```python
moviedata['director'][0]
```

```python
moviedata['tags'] = moviedata['overview'] + moviedata['genres'] + moviedata['keywords'] + moviedata['cast'] + moviedata['director']
```

```python
moviedata.head()
```

```python
new = moviedata.drop(columns=['overview','genres','keywords','cast','director'])
```

```python
new.head()
```

 **Text vectorization**

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
new[new['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```
